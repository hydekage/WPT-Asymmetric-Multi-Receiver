First, we wrote a parameter symbol for the WPT circuit which is $R_S,R_1,C,L_1,M$ and $R_L,R_2,L_2$ using KCL & KVL form as: 

Primary Circuit-Voltage Law:
$$u = {i_1}{R_S}+{i_1}{R_1}+{L_1}{\frac{di_1}{dt}}+{M_1}{\frac{di_2}{dt}}+{M_2}{\frac{di_3}{dt}}+v_1$$
Primary Circuit-Current Law:
$$i_1=C\frac{dv_1}{dt}$$

1st receiver-Voltage Law:
$$0 = {i_2}{R_{L1}}+{i_2}{R_{2}}+{L_2}{\frac{di_2}{dt}}+{M_1}{\frac{di_1}{dt}}+{M_3}{\frac{di_3}{dt}}$$

2nd receiver-Voltage Law:
$$0 = {i_3}{R_{L2}}+{i_3}{R_{3}}+{L_3}{\frac{di_3}{dt}}+{M_2}{\frac{di_1}{dt}}+{M_3}{\frac{di_2}{dt}}$$



```python
i1d = (sp.Function('i_1')(t).diff(t))
i2d = (sp.Function('i_2')(t).diff(t))
i3d = (sp.Function('i_3')(t).diff(t))
v1d = (sp.Function('v_1')(t).diff(t))
eq1 = u-i1*RS-i1*R1-(L1*i1d+M1*i2d+M2*i3d)-v1
eq2 = -(i2*R2)-(i2*RL1)-(L2*i2d+M1*i1d+M3*i3d)
eq3 = -(i3*R3)-(i3*RL2)-(L3*i3d+M2*i1d+M3*i2d)
eq4 = C*v1d-i1
display('KVL Primary:',eq1,'KCL Primary:',eq4,'KVL 1st receiver:',eq2,'KVL 2nd Receiver:',eq3)
```

In the Wireless Power Transfer(WPT) circuit, find the state space variables and afterwards we will obtain matrix A and B in the state space equation. Therefore, we need to define the matrix A and B using below operation in sympy. the equation:
$$x ̇=Ax+Bu$$
$$y=Cx+Du$$  
define the WPT state space equation as:
$$
\begin{bmatrix}
\frac{dv_1}{dt}\\
\frac{di_1}{dt}\\
\frac{di_2}{dt}\\
\frac{di_3}{dt}\\
\end{bmatrix}
= A
\begin{bmatrix}
v_1\\
i_1\\
i_2\\
i_3\\
\end{bmatrix}
+ B
u
$$

$$
y = 
C
\begin{bmatrix}
v_1\\
i_1\\
i_2\\
i_3\\
\end{bmatrix}+ Du
$$

```python
sol = sp.solve((eq1,eq2,eq3,eq4),(i1d,i2d,i3d,v1d),Dict=True)
print(sol)
solv1d = sb.named_expression(r'\frac{dv_1}{dt}',sol[v1d])
display(solv1d.as_eq())
soli1d = sb.named_expression(r'\frac{di_1}{dt}',sol[i1d])
display(soli1d.as_eq())
soli2d = sb.named_expression(r'\frac{di_2}{dt}',sol[i2d])
display(soli2d.as_eq())
soli3d = sb.named_expression(r'\frac{di_3}{dt}',sol[i3d])
display(soli3d.as_eq())
```

```python
dv1dt = sp.poly(solv1d,(v1,i1,i2,i3,u)).as_dict()
di1dt = sp.poly(soli1d,(v1,i1,i2,i3,u)).as_dict()
di2dt = sp.poly(soli2d,(v1,i1,i2,i3,u)).as_dict()
di3dt = sp.poly(soli3d,(v1,i1,i2,i3,u)).as_dict()
#print("display i1 coefficient on dv1/dt:",dv1dt[0,1,0,0,0])
#print(dv1dt,"\n\n",di1dt,"\n\n",di2dt,"\n\n",di3dt,"\n\n")
```

### 2.2.1 State Space Matrix definition: ###

```python
A_Row_1 = (0,dv1dt[0,1,0,0,0],0,0)
A_Row_2 = (di1dt[1,0,0,0,0],di1dt[0,1,0,0,0],di1dt[0,0,1,0,0],di1dt[0,0,0,1,0])
A_Row_3 = (di2dt[1,0,0,0,0],di2dt[0,1,0,0,0],di2dt[0,0,1,0,0],di2dt[0,0,0,1,0])
A_Row_4 = (di3dt[1,0,0,0,0],di3dt[0,1,0,0,0],di3dt[0,0,1,0,0],di3dt[0,0,0,1,0])
A  = Matrix([A_Row_1,A_Row_2,A_Row_3,A_Row_4])
display('A:',A)
B_Row_1 = (0)
B_Row_2 = (di1dt[0,0,0,0,1])
B_Row_3 = (di2dt[0,0,0,0,1])
B_Row_4 = (di3dt[0,0,0,0,1])
B = Matrix([B_Row_1,B_Row_2,B_Row_3,B_Row_4])
display('B:',B)
```

```python
display(delta)
deltav = (L1*L2*L3 - L1*M3**2 - L2*M2**2 - L3*M1**2 + 2*M1*M2*M3)
A_simp = A.subs([(deltav,delta),(1/C,delta/C)])
display(deltav)
display(A_simp)
```

```python
CL1 = Matrix([[0,0,-RL1,0]])
CL2 = Matrix([[0,0,0,-RL2]])
CS = Matrix([[0,-RS,0,0]])
D = Matrix([[0]])
display("CL1:",CL1,"CL2:",CL2,"CS:",CS,"D:",D)
```

### 2.2.2 Transfer Function Calculation: ###
$$\frac{Y(s)}{U(s)} =C[sI−A]^{−1} B+D$$
$$G_S(s)=\frac{Y_S(s)}{U(s)}$$
$$G_L(s)=\frac{Y_L(s)}{U(s)}$$

```python
I = sp.eye(A.shape[0])
ISA = sp.inv_quick(I*s - A)
#ISA = (I*s-A).inv()
TFL1 = CL1*ISA*B + D
TFL2 = CL2*ISA*B + D
TFS = CS*ISA*B + D
#TFL1 = CL1*(ISA)*B + D
#TFL2 = CL2*(ISA)*B + D
#TFS = CS*((I*s - A).inv())*B + D
GLS1 =TFL1[0].simplify().collect(s)
GLS2 =TFL2[0].simplify().collect(s)
GSS =TFS[0].simplify().collect(s)
GLS1e = sb.named_expression('G_{L1}(s)',GLS1)
GLS2e = sb.named_expression('G_{L2}(s)',GLS2)
GSSe = sb.named_expression('G_S(s)',GSS)
display(GLS1e.as_eq())
display(GLS2e.as_eq())
display(GSSe.as_eq())
```

```python
GLS1n,GLS1d =sp.fraction(GLS1) 
GLS2n,GLS2d =sp.fraction(GLS2)
deltav = sb.named_expression(r'\Delta',GLS1d) 
GLS1_simp=sb.named_expression('G_{L1}(s)',GLS1.subs(deltav,delta))
GLS2_simp=sb.named_expression('G_{L2}(s)',GLS2.subs(deltav,delta))
display(GLS1_simp.as_eq(),GLS2_simp.as_eq(),deltav.as_eq().simplify())
```

```python
GLS1S = GLS1.subs(s,sp.I*omega)
GLS2S = GLS2.subs(s,sp.I*omega)
#GLS1A = GLS1S 
#GLS2A = GLS2S
GLS1A = sp.Abs(GLS1S).simplify()**2/RL1
GLS2A = sp.Abs(GLS2S).simplify()**2/RL2
GLS1Ae = sb.named_expression('|G_{L1}(j\omega)|',GLS1A)
GLS2Ae = sb.named_expression('|G_{L2}(j\omega)|',GLS2A)
#display(GLS1Ae.as_eq())
#display(GLS2Ae.as_eq())
```

```python
class MyProblem(Problem):
    def __init__(self):
        super().__init__(n_var=3,
                         n_obj=2,
                         n_constr=0,
                         xl=np.array([1e-12,100,0.01]),
                         xu=np.array([1e-1,1e9,0.2]),
                         elementwise_evaluation=True)

    def _evaluate(self, x, out, *args, **kwargs):
        f1 = -(x[0]**2*x[1]**4*x[2]**2*(1.56816e-19*x[1]**2 + 4.361809936e-7)/(x[0]**2*x[1]**2*(x[1]**6*(1.8974736e-29*x[2]**4 - 1.8974736e-29*x[2]**2 + 4.743684e-30) + x[1]**4*(2.723169856e-17*x[2]**4 - 3.03760396016e-17*x[2]**2 + 1.57198056169e-17) + x[1]**2*(1.2673656827568e-6*x[2]**2 + 7.02946473350086e-6) + 14745.8280126248) + x[0]*x[1]**2*(x[1]**4*(8.62488e-25*x[2]**2 - 4.31244e-25) + x[1]**2*(1.4281643288e-12*x[2]**2 - 1.4281643288e-12) - 0.63603218562444) + 9.801e-21*x[1]**4 + 3.24582802e-8*x[1]**2 + 14455.27694601))
        f2 = -(x[0]**2*x[1]**4*x[2]**2*(3.5937e-19*x[1]**2 + 1.90555497e-7)/(x[0]**2*x[1]**2*(x[1]**6*(1.8974736e-29*x[2]**4 - 1.8974736e-29*x[2]**2 + 4.743684e-30) + x[1]**4*(2.723169856e-17*x[2]**4 - 3.03760396016e-17*x[2]**2 + 1.57198056169e-17) + x[1]**2*(1.2673656827568e-6*x[2]**2 + 7.02946473350086e-6) + 14745.8280126248) + x[0]*x[1]**2*(x[1]**4*(8.62488e-25*x[2]**2 - 4.31244e-25) + x[1]**2*(1.4281643288e-12*x[2]**2 - 1.4281643288e-12) - 0.63603218562444) + 9.801e-21*x[1]**4 + 3.24582802e-8*x[1]**2 + 14455.27694601))
        out["F"] = [f1, f2]
        #out["G"] = [g1, g2]
```
